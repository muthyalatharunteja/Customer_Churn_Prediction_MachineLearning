# FEATURE ENGINEERING

```python
df.head(1)
```

```python
df.columns
```

```python
y=df[['Churn']]
X=df[['Age','Gender','Tenure','MonthlyCharges']]
```

```python
X  # 1-Female & 0-male
```

```python
X['Gender']= X['Gender'].apply(lambda x: 1 if x=='Female' else 0)

# Encode categorical variables to numeric values.

# if lambda(x) = 1 then x=female, if  0 then x=male
```

```python
 X.head()
```

```python
type(X['Gender'][0])
```

```python
y
```

```python
y['Churn']= y['Churn'].apply(lambda x: 1 if x=='Yes'else 0)
# if lambda(x) = 1 then x=yes, if  0 then x=no
```

```python
y
```

```python
type(y['Churn'][0])
```

# TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
#testing 20% and training 80%
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler

## PREPROCESSING
# Standardize features to have mean 0 and std 1.
# Scaling helps algorithms like KNN and SVM perform better. We save the scaler for later use in production.
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
import joblib
joblib.dump(scaler,'scaler.pkl')

# scaler exported as scaler.pkl
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
X_train
```

```python
# X_test
```

# Modelling

### Logistic Regression

```python
from sklearn.metrics import accuracy_score
def modelperformance(predictions):
    print('Accuracy score model is {}'.format(accuracy_score(y_test,predictions)))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model= LogisticRegression()
```

```python
log_model.fit(X_train, y_train)
```

```python
X
```

```python
log_model.predict([[40,0,20,80]])[0]
```

```python
y_pred= log_model.predict(X_test)
```

```python
y_test
```

```python
modelperformance(y_pred)
#predicting customer churn or not
```

### KNN Neighbours

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
param_grid={
    'n_neighbors':[3,5,7,9],
    'weights':['uniform','distance'],
    }
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
gridkn = GridSearchCV(KNeighborsClassifier(),param_grid,cv=5)
```

```python
gridkn.fit(X_train, y_train)
```

```python
gridkn.best_params_
```

```python
y_pred= gridkn.predict(X_test)
```

```python
modelperformance(y_pred)
```

```python
from sklearn.svm import SVC
```

```python
svm= SVC()
```

```python
param_grid={
    'C':[0.01,0.1,0.5,1],
    'kernel':['linear','rbf','poly']
}
```

```python
gridsvc = GridSearchCV(svm,param_grid,cv=5)
#support vector classifier
```

```python
gridsvc.fit(X_train, y_train)
```

```python
gridsvc.best_params_
```

```python
y_pred=gridsvc.predict(X_test)
```

```python
modelperformance(y_pred)
```

## DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
param_grid={
    "criterion": ["gini", "entropy"],
    #tree decides the best split. "gini" = Gini impurity, "entropy" = information gain
    "splitter": ["best", "random"],
    "max_depth": (None, 5, 20, 30), # depth of tree
    "min_samples_split": [2,5,10],
    "min_samples_leaf": [1,2,4]
        }
```

```python
grid_tree = GridSearchCV(DecisionTreeClassifier(),param_grid,cv=5)
```

```python
grid_tree.fit(X_train,y_train)
```

```python
y_pred=grid_tree.predict(X_test)
```

```python
y_pred
```

```python
modelperformance(y_pred)
```

## Random Forest classifer

```python
from sklearn.ensemble import RandomForestClassifier

rfc_model = RandomForestClassifier()
param_grid = {
    'n_estimators':[32,64,128,256],
    'max_features' :[2,3,4],
    "bootstrap":[True, False]
}
```

```python
grid_rfc = GridSearchCV(rfc_model,param_grid,cv=5)
```

```python
grid_rfc.fit(X_train,y_train)
```

```python
grid_rfc.best_params_
```

```python
y_pred=grid_rfc.predict(X_test)
```

```python
y_pred
```

```python
modelperformance(y_pred)
```
